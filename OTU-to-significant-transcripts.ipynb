```python
class Findsig(object):
    """
    This class uses as input a file containing OTU-transcript-r square value, uses as query a term contained in the 
    OTU part and returns a list of transcripts sorted by their respective r square value
    """
    
    version = 0.1
    
    def __init__(self, inp):
        self.inp = inp

    def findsig(self, query):
        match = [ i for i in self.inp if query in i]
        srtd = sorted(match, key=lambda x:(float(x.split(',')[-1])))[::-1]   
        return [i for i in srtd if float(i.split(',')[-1]) > p99]
```
